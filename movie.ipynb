### Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

### Dataset

```python
df = pd.read_csv("assets/movies.csv")

# menampilkan 5 baris pertama
print("Lima baris pertama:")
df.head()
```

```python
# ringkasan informasi data
print("\nRingkasan informasi data:")
df.info()
```

## Pre-processing

```python
# memeriksa dan menampilkan jumlah nilai null per kolom
null_counts = df.isnull().sum()

# menampilkan hasil
print("Jumlah data null per kolom:")
print(null_counts)
```

```python
# menghapus baris yang mengandung nilai null
df_cleaned = df.dropna()

# menghitung jumlah baris awal dan akhir
initial_row_count = df.shape[0]
final_row_count = df_cleaned.shape[0]

# menghitung jumlah baris yang dihapus
rows_deleted = initial_row_count - final_row_count

# menampilkan hasil
print(f"\nJumlah baris awal: {initial_row_count}")
print(f"Jumlah baris setelah pembersihan: {final_row_count}")
print(f"Jumlah baris yang dihapus: {rows_deleted}")
```

```python
# mengecek duplikat data
duplicates = df_cleaned[df_cleaned.duplicated()]

# menampilkan hasil
print("\nBaris duplikat:")
print(duplicates)
print(f"Jumlah baris duplikat: {duplicates.shape[0]}")
```

```python
# visualisasi hubungan antar variabel sebelum pembersihan
sns.pairplot(df)
plt.suptitle('Hubungan Antar Variabel Sebelum Pembersihan', y=1, ha='center')
plt.show()

# visualisasi hubungan antar variabel setelah pembersihan
sns.pairplot(df_cleaned)
plt.suptitle('Hubungan Antar Variabel Setelah Pembersihan', y=1, ha='center')
plt.show()
```

Setelah dilakukan pembersihan data dengan menghapus baris yang mengandung nilai null, visualisasi distribusi dan hubungan antar variabel tidak menunjukkan perubahan drastis. Hal ini mengindikasikan bahwa pembersihan data tidak mempengaruhi pola atau tren yang ada dalam data.
